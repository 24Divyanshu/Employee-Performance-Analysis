```python
sns.lineplot(x='Age',y='TotalWorkExperienceInYears',data=df)

plt.title('Relation b/w Age & TotalWorkExperienceInYears')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='EmpJobRole',hue=df.PerformanceRating,data=df)
plt.xticks(rotation='vertical')
plt.xlabel('EmpJobRole',fontsize=20)
plt.ylabel('PerformanceRating',fontsize=15)
plt.show()
```
